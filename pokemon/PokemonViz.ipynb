```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pokemon = pd.read_csv("./Pokemon.csv")
pokemon.head()
```

```python
%matplotlib inline
```

```python
sns.set(style="white", color_codes=True)
```

```python
pokemon.plot(kind="scatter", x="Attack", y="Defense")
```

```python
sns.jointplot(x="Attack", y="Defense", data=pokemon)
```

```python
sns.FacetGrid(pokemon, hue="Type 1", size=8) \
    .map(plt.scatter, "Attack", "Defense") \
    .add_legend()
```

```python
sns.FacetGrid(pokemon, hue="Generation", size=8) \
    .map(plt.scatter, "Attack", "Defense") \
    .add_legend()
```

```python
from pandas.tools.plotting import andrews_curves
andrews_curves(pokemon.drop(["#", "Name", "Type 2"], axis=1), "Type 1")
```

```python
from pandas.tools.plotting import parallel_coordinates
parallel_coordinates(pokemon.drop(["#", "Name", "Type 2"], axis=1), "Type 1")
```

```python
from pandas.tools.plotting import radviz
radviz(pokemon.drop(["#", "Name", "Type 2", "Legendary"], axis=1), "Type 1")
```

```python
legendary = pokemon.query("Legendary == True")
```

```python
non_legendary = pokemon.query("Legendary == False")
```

```python
f, ax = plt.subplots(figsize=(6, 6))
ax.set_aspect("equal")

ax = sns.kdeplot(legendary.Attack, legendary.Defense,
                cmap="Reds", shade=True, shade_lowest=False)
ax = sns.kdeplot(non_legendary.Attack, non_legendary.Defense,
                cmap="Blues", shade=True, shade_lowest=False)
ax.figure.suptitle("Legendary vs Non-Legendary")
```
